```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt


from utils import readPd, PlotBode
```

```python
df = readPd('C_HP.csv')
# reinterpret as float
df = df.astype(float)
df.info()
```

```python
f = df['Frecuencia'].values
Z_s = df['|Z|[uS]'].values * 1e-6
Z_mod = (1/Z_s)
Z_phase = df['Fase'].values

plot = PlotBode()
plot.plotLoglog1(f, Z_mod, lw=0, marker='.', color='r', label='|Z|')
plot.plotLoglog2(f, Z_phase, lw=0, marker='.', color='b', label='Fase')
plot.show(min2loc=0.5, maj2loc=1)
```
